### m. Quels sont les couples de sportifs (identifiant et nom et prénom de chaque) de même age ?

```python
# Récupérer tous les sportifs
sportifs = list(db.Sportifs.find({}, {"IdSportif": 1, "Nom": 1, "Prenom": 1, "Age": 1}))

# Dictionnaire pour stocker les sportifs par âge
sportifs_par_age = {}

# Grouper les sportifs par âge
for sportif in sportifs:
    age = sportif['Age']
    if age not in sportifs_par_age:
        sportifs_par_age[age] = []
    sportifs_par_age[age].append(sportif)

# Trouver les couples
couples = []

for age, sportifs in sportifs_par_age.items():
    if len(sportifs) > 1:  # Si plus d'un sportif partage le même âge
        for s1, s2 in combinations(sportifs, 2):
            couples.append((s1['IdSportif'], s1['Nom'], s1['Prenom'], s2['IdSportif'], s2['Nom'], s2['Prenom']))

# Affichage des couples
for couple in couples:
    print(f"Sportif 1: ID={couple[0]}, Nom={couple[1]}, Prénom={couple[2]} | "
          f"Sportif 2: ID={couple[3]}, Nom={couple[4]}, Prénom={couple[5]}")
```

### r. Quels sont les couples de sportifs (identifiant et nom et prénom de chaque) de même âge avec le même conseiller ?

```python
from itertools import combinations

# Récupérer tous les sportifs avec leur âge et conseiller
sportifs = list(db.Sportifs.find({}, {"IdSportif": 1, "Nom": 1, "Prenom": 1, "Age": 1, "IdSportifConseiller": 1}))

# Dictionnaire pour stocker les sportifs par âge et conseiller
sportifs_par_age_et_conseiller = {}

# Grouper les sportifs par âge et conseiller
for sportif in sportifs:
    age = sportif['Age']
    conseiller = sportif.get('IdSportifConseiller')  # Obtenir l'ID du conseiller
    if conseiller is not None:  # Filtrer les sportifs qui ont un conseiller
        key = (age, conseiller)  # Créer une clé basée sur l'âge et le conseiller
        if key not in sportifs_par_age_et_conseiller:
            sportifs_par_age_et_conseiller[key] = []
        sportifs_par_age_et_conseiller[key].append(sportif)

# Trouver les couples
couples = []

for key, sportifs in sportifs_par_age_et_conseiller.items():
    if len(sportifs) > 1:  # Si plus d'un sportif partage le même âge et le même conseiller
        for s1, s2 in combinations(sportifs, 2):  # Trouver les paires
            couples.append((s1['IdSportif'], s1['Nom'], s1['Prenom'], s2['IdSportif'], s2['Nom'], s2['Prenom']))

# Affichage des couples
print("Couples de sportifs de même âge et ayant le même conseiller :")
for couple in couples:
    print(f"Sportif 1: ID={couple[0]}, Nom={couple[1]}, Prénom={couple[2]} | "
          f"Sportif 2: ID={couple[3]}, Nom={couple[4]}, Prénom={couple[5]}")
```

### t. Pour chaque sportif donner le nombre de sports qu’il arbitre.

```python
# Compter le nombre de sport que chaque sportifs dans Sportifs.Sports.Arbitrer
query = [
    {
        "$unwind": "$Sports.Arbitrer"
    },
    {
        "$group": {
            "_id": "$IdSportif",
            "count": {"$sum": 1}
        }
    }
]

result = db.Sportifs.aggregate(query)

for i in result:
    print(i)
```

### u. Pour chaque gymnase de Stains donner par jour d’ouverture les horaires des premières et dernières séances.

```python
# Pour chaque gymnase de Stains donner par jour d’ouverture les horaires des premières et dernières séances.
query = [
    {
        "$match": {
            "Ville": "STAINS"
        }
    },
    {
        "$unwind": "$Seances"
    },
    {
        "$group": {
            "_id": "$NomGymnase",
            "first": {"$min": "$Seances.Horaire"},
            "last": {"$max": "$Seances.Horaire"}
        }
    }
]

result = db.Gymnases.aggregate(query)

for i in result:
    print(i)
```

### v. Pour chaque entraîneur de hand ball quel est le nombre de séances journalières qu’il assure ?

```python
pipeline = [
    {"$match": {"Sports.Entrainer": "Hand ball"}},
    {"$lookup": {
        "from": "Gymnases",
        "let": {"entraineurId": "$IdSportif"},
        "pipeline": [
            {"$unwind": "$Seances"},
            {"$match": {
                "$expr": {
                    "$and": [
                        {"$eq": ["$Seances.IdSportifEntraineur", "$$entraineurId"]},
                        {"$eq": ["$Seances.Libelle", "Hand ball"]}
                    ]
                }
            }}
        ],
        "as": "seances"
    }},
    {"$unwind": "$seances"},
    {"$group": {
        "_id": {
            "entraineur": "$Nom",
            "jour": "$seances.Seances.Jour"
        },
        "nombreSeances": {"$sum": 1}
    }},
    {"$group": {
        "_id": "$_id.entraineur",
        "seancesParJour": {
            "$push": {
                "jour": "$_id.jour",
                "nombre": "$nombreSeances"
            }
        }
    }}
]
#
result = db.Sportifs.aggregate(pipeline)

print("Nombre de séances journalières pour chaque entraîneur de handball ")
for entraineur in result:
    print(f"Entraîneur: {entraineur['_id']}")
    for seance in entraineur['seancesParJour']:
        print(f"  {seance['jour']}: {seance['nombre']} séance(s)")
    print()
```

```python
pipeline = [
    {"$unwind": "$Seances"},  # Éclater les séances pour chaque gymnase
    # {"$match": {"Seances.Jour": "mercredi"}},  # Filtrer les séances qui se déroulent le mercredi
    {"$group": {
        "_id": "$IdGymnase",  # Grouper par gymnase
        "nombreSeancesMercredi": {"$sum": 1}  # Compter le nombre de séances par gymnase pour le mercredi
    }},
    {"$match": {"nombreSeancesMercredi": {"$gt": 15}}},  # Ne garder que les gymnases ayant plus de 15 séances le mercredi
    {"$lookup": {  # Joindre avec la collection des gymnases pour obtenir leurs noms
        "from": "Gymnases",
        "localField": "_id",
        "foreignField": "IdGymnase",
        "as": "gymnaseDetails"
    }},
    {"$unwind": "$gymnaseDetails"},  # Éclater les détails des gymnases
    {"$project": {  # Projeter uniquement les champs pertinents
        "_id": 0,
        "NomGymnase": "$gymnaseDetails.NomGymnase",
        "nombreSeancesMercredi": 1
    }}
]

# Exécution de l'agrégation
result = db.Gymnases.aggregate(pipeline)
for i in result:
    print(i)
```
